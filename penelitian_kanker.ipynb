```python
#import library

import pandas as pd
```

Fit Transform

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop_duplicates()
```

```python
dataset.duplicated().any()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
d=dataset.apply(le.fit_transform)
```

```python
#Splitting

X =d.loc[:,"Age (years)":"Ethnicity"]
y = d["Grouping"]
```

```python
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test,y_train, y_test = train_test_split(X,y, random_state=100,  

                                   test_size=0.20,  

                                   shuffle=True)
```

```python
dataset.head()
```

Modeling and HyperParameter Tuning

```python
# Necessary imports KNN
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.svm import SVC
```

```python
knn_clf = KNN()
svc_clf = SVC()
```

```python
knn_clf.fit(X_train, y_train)
```

```python
knn_clf.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = [ { 'n_neighbors': list(range(1,9)),
                        'algorithm': ['auto', 'ball_tree', 'kd_tree' , 'brute'] } ]

# defining parameter range
param_grid2 = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
```

```python
gs = GridSearchCV(knn_clf,param_grid,cv=20)
gs.fit(X_train, y_train)
```

```python
gs1 = GridSearchCV(svc_clf,param_grid2,refit = True, verbose = 100)
gs1.fit(X_train, y_train)
```

```python
pd.DataFrame(gs.cv_results_['params'])
```
